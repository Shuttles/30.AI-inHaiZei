```python
train_X = np.array([30.0, 40.0, 60.0, 80.0, 100.0, 120.0, 140.0])
train_Y = np.array([320.0, 360.0, 400.0, 455.0, 490.0, 560.0, 580.0])
train_X /= 100.0
train_Y /= 100.0

plot_points(train_X, train_Y, title_name='Training Points')
```

```python
x = tf.placeholder(tf.float32, None)
y = tf.placeholder(tf.float32, None)
w = tf.Variable(np.random.randn(), name='weight')
b = tf.Variable(np.random.randn(), name='bias')
y_pred = tf.add(tf.multiply(w, x), b)#y预测值
#print(w)
#print(b)
#print(y_pred)
```

```python
n_samples = train_X.shape[0]
cost = tf.reduce_sum(tf.pow((y_pred - y), 2)) / n_samples
learning_rate = 0.01
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```python
training_epoch = 1000
display_step = 50
```

```python
with tf.Session() as sess:
    sess.run(tf.initialize_all_variables()) #初始化变量
    for epoch in range(training_epoch):
        for (x_train, y_train) in zip(train_X, train_Y):
            #优化
            _, loss = sess.run([optimizer, cost], feed_dict={x : x_train, y : y_train})
            print(loss)
```
